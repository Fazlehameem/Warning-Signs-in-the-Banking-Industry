```python
# Initial imports:
import pandas as pd
from pathlib import Path

import hvplot.pandas
from sqlalchemy import create_engine
```

```python
# Set path for "Failed Banks 2008-2010" csv files:
csvpath_first_georgia_community_bank = Path("Failed_Banks_2008_2010/first_georgia_community_bank.csv")
csvpath_security_pacific_bank = Path("Failed_Banks_2008_2010/security_pacific_bank.csv")
csvpath_haven_trust_bank = Path("Failed_Banks_2008_2010/haven_trust_bank.csv")
csvpath_firstbank_financial_services = Path("Failed_Banks_2008_2010/firstbank_financial_services.csv")
csvpath_bank_of_clark_county = Path("Failed_Banks_2008_2010/bank_of_clark_county.csv")
csvpath_first_centennial_bank = Path("Failed_Banks_2008_2010/first_centennial_bank.csv")
csvpath_liberty_bank = Path("Failed_Banks_2008_2010/liberty_bank.csv")
csvpath_first_national_bank_of_the_south = Path("Failed_Banks_2008_2010/first_national_bank_of_the_south.csv")
csvpath_peninsula_bank = Path("Failed_Banks_2008_2010/peninsula_bank.csv")
csvpath_first_national_bank = Path("Failed_Banks_2008_2010/first_national_bank.csv")
csvpath_first_georgia_community_bank_two = Path("Failed_Banks_2008_2010/first_georgia_community_bank_two.csv")
csvpath_first_georgia_community_bank_three = Path("Failed_Banks_2008_2010/first_georgia_community_bank_three.csv")
```

```python
# Read in data for "Failed Banks 2008-2010" csv files:
first_georgia_community_bank_df = pd.read_csv(csvpath_first_georgia_community_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
security_pacific_bank_df = pd.read_csv(csvpath_security_pacific_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
haven_trust_bank_df = pd.read_csv(csvpath_haven_trust_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
firstbank_financial_services_df = pd.read_csv(csvpath_firstbank_financial_services, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
bank_of_clark_county_df = pd.read_csv(csvpath_bank_of_clark_county, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
first_centennial_bank_df = pd.read_csv(csvpath_first_centennial_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
liberty_bank_df = pd.read_csv(csvpath_liberty_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
first_national_bank_of_the_south_df = pd.read_csv(csvpath_first_national_bank_of_the_south, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
peninsula_bank_df = pd.read_csv(csvpath_peninsula_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
first_national_bank_df = pd.read_csv(csvpath_first_national_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
first_georgia_community_bank_two_df= pd.read_csv(csvpath_first_georgia_community_bank_two, index_col='Quarter', parse_dates=True, infer_datetime_format=True)
first_georgia_community_bank_three_df= pd.read_csv(csvpath_first_georgia_community_bank_three, index_col='Quarter', parse_dates=True, infer_datetime_format=True)
```

```python
# Clean "Failed Banks 2008-2010" data by dropping nulls, dropping duplicates, and renaming columns:
first_georgia_community_bank_df.dropna(inplace=True)
security_pacific_bank_df.dropna(inplace=True)
haven_trust_bank_df.dropna(inplace=True)
firstbank_financial_services_df.dropna(inplace=True)
bank_of_clark_county_df.dropna(inplace=True)
first_centennial_bank_df.dropna(inplace=True)
liberty_bank_df.dropna(inplace=True)
first_national_bank_of_the_south_df.dropna(inplace=True)
peninsula_bank_df.dropna(inplace=True)
first_national_bank_df.dropna(inplace=True)
first_georgia_community_bank_two_df.dropna(inplace=True)
```

```python
# Set path for "Failed Banks Post 2010" csv files:
csvpath_community_south_bank_and_trust = Path("Failed_Banks_Post_2010/community_south_bank_and_trust.csv")
csvpath_community_south_bank = Path("Failed_Banks_Post_2010/community_south_bank.csv")
csvpath_harvest_community_bank = Path("Failed_Banks_Post_2010/harvest_community_bank.csv")
csvpath_nova_bank = Path("Failed_Banks_Post_2010/nova_bank.csv")
csvpath_douglas_county_bank = Path("Failed_Banks_Post_2010/douglas_county_bank.csv")
```

```python
# Read in data for "Failed Banks Post 2010" csv files:
community_south_bank_and_trust_df = pd.read_csv(csvpath_community_south_bank_and_trust, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
community_south_bank_df = pd.read_csv(csvpath_community_south_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
harvest_community_bank_df = pd.read_csv(csvpath_harvest_community_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
nova_bank_df = pd.read_csv(csvpath_nova_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
douglas_county_bank_df = pd.read_csv(csvpath_douglas_county_bank, index_col='bank_name', parse_dates=True, infer_datetime_format=True)
```

```python
# Clean "Failed Banks Post 2010" data by dropping nulls, dropping duplicates, and renaming columns:
community_south_bank_and_trust_df.dropna(inplace=True)
community_south_bank_df.dropna(inplace=True)
harvest_community_bank_df.dropna(inplace=True)
nova_bank_df.dropna(inplace=True)
douglas_county_bank_df.dropna(inplace=True)
```
